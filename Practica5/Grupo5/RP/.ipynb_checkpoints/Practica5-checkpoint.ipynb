```python
import matplotlib
import imageio
import matplotlib.pyplot as plt
from scipy import ndimage
from scipy import signal
import numpy as np 
import pandas as pd
```

```python
import os
import re 

files = {os.path.splitext(k)[0]: {} for k in os.listdir('data') if 'ej' in os.path.splitext(k)[0]}


for i,file in enumerate(os.listdir('data')):
    name = os.path.splitext(file)[0]
    if 'ej' in name:
        df = pd.read_csv(f'data/{file}')
        files[name] = df
```

```python
files.keys()
```

```python
import seaborn as sns
plt.style.use('seaborn-darkgrid')
```

## Ejercicio 4
### (a) Barrido de frecuencias

```python
df4a = files['data_ej_4_a']
df4a.columns
```

```python
plt.figure(figsize=(10,5))
plt.semilogx(df4a['frec'],df4a['X'],'.')
plt.ylabel(r'Salida [V]',fontsize=20)
plt.xlabel(r'Log( Frecuencia [Hz] )',fontsize=20)
plt.tick_params(axis="x", labelsize=20)
plt.tick_params(axis="y", labelsize=20)
plt.title("Barrido de frecuencias de Lock-in",fontsize=20)
```

### (b) Lock in en funcion de frecuencias

```python
df4b = files['data_ej_4_b']
df4b.columns
```

```python
plt.figure(figsize=(10,5))
plt.semilogy(df4b['amp'],df4b['X'])
plt.ylabel(r'Log (Salida [V] )',fontsize=20)
plt.xlabel(r'Amplitud de Modulación [V] ',fontsize=20)
plt.tick_params(axis="x", labelsize=20)
plt.tick_params(axis="y", labelsize=20)
plt.title("Respuesta del Lock-in a 102Hz en función de la amplitud",fontsize=20)
```
